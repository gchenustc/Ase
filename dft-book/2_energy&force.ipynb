```python
# 计算co的能量和力
```

```python
from ase import Atoms,Atom
from ase.calculators.vasp import Vasp

co = Atoms([
    Atom("C", [0,0,0]),
    Atom("O", [1.2,0,0]),
], cell=[6,6,6],pbc=True) # pbc必须要打开，否则会报错

calc = Vasp(atoms=co, 
            xc="pbe",
            encut=350,
            ismear=1,
            sigma=0.01,
            txt="vasp.out")

# print(co.get_potential_energy())
# print(co.get_forces())
# Vasp 类对象具有的属性和方法
dir(calc)[49:]
```

```python
# 计算co能量和胞长的关系
```

```python
from ase import Atoms,Atom
from ase.calculators.vasp import Vasp

co = Atoms([
    Atom("C", [0,0,0]),
    Atom("O", [1.2,0,0]),
])



L = [4, 5, 6, 8, 10]
energies = []

for a in L:
    co.set_cell([a,a,a],scale_atoms=False)
    co.center()
    co.pbc=True # 这一步必须有

    calc = Vasp(xc="pbe",
                encut=350,
                ismear=1,
                sigma=0.01,
                txt=f"{a}.out")
    
    co.calc = calc
    energies.append(co.get_potential_energy())

import matplotlib.pyplot as plt
plt.plot(L, energies, 'bo-')
plt.xlabel("Unit cell length ($\AA$)")
plt.ylabel("Total energy (eV)")
plt.savefig("./images/co-e-v.png")
```
